#começo

estou importando o pandas e apelidei de pd para não ter que escrever pandas total hora

```python
import pandas as pd
```

**estou fazendo um dicionario, que vai acabar virando um dataframe**


```python
relatorio = {
    'Dia': ['segunda' ,'terça' ,'quarta' ,'quinta' ,'sexta' ,'sabado' ,'domingo'],
    'Horas Trabalhadas':[6 ,7 ,8 ,6 ,7 ,5 ,4],
    'Bugs Corrigidos': [3 ,2 ,1 ,4 ,3 ,2 ,1],
    'Tarefas concluidas': [5 ,4 ,6 ,4 ,5 ,3 ,2],
    }
```

**estou colocando a função re, para a cada momento eu chamar ela,ela rodar**

```python
re = pd.DataFrame(relatorio)
```

**estou fazendo a minha tabela roda**

```python
re
```

# finalizando o trabalho

**estou criando a tabela total de horas trabalhadas**


```python
re['Total de Horas Trabalhadas'] = re['Horas Trabalhadas'].sum()
```

**estou criando a tabela Média Diária de Horas Trabalhadas**


```python
re['Média Diária de Horas Trabalhadas'] = re['Horas Trabalhadas'].median()
```

**estou criando a tabela Total de Bugs Corrigidos**

```python
re['Total de Bugs Corrigidos'] = re['Bugs Corrigidos'].sum()
```

**estou criando a tabela Média Diária de Bugs Corrigidos**

```python
re['Média Diária de Bugs Corrigidos'] = re['Bugs Corrigidos'].median()
```

**estou criando a tabela Total de Tarefas Concluídas**

```python
re['Total de Tarefas Concluídas'] = re['Tarefas concluidas'].sum()
```

**estou criando a tabela Média Diária de Tarefas Concluídas**

```python
re['Média Diária de Tarefas Concluídas'] = re['Tarefas concluidas'].median()
```

**estou criando a tabela Produtividade Diária (Tarefas Concluídas por Hora)**

```python
re['Produtividade Diária (Tarefas Concluídas por Hora)'] = re['Horas Trabalhadas'] / re['Tarefas concluidas']
```
